Contexto


Você trabalha em uma plataforma de streaming e a diretoria está preocupada com o
alto índice de usuários cancelando as suas assinaturas. Eles acreditam que é possível

```python
X = df.drop('Churned_label', axis=1)
y = df['Churned_label']
```

```python
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
predictions = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, predictions)
print(f"Acurácia: {accuracy*100:.2f}")
```

```python

predictions = rf.predict(X_test)


f1 = f1_score(y_test, predictions, average='weighted')


recall = recall_score(y_test, predictions, average='weighted')


precision = precision_score(y_test, predictions, average='weighted')

print(f"F1 Score: {f1*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
print(f"Precisão: {precision*100:.2f}%")
```

```python
scaler = StandardScaler(with_mean=False)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)


predictions = model.predict(X_test_scaled)


f1 = f1_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')
precision = precision_score(y_test, predictions, average='weighted')

print(f"F1 Score: {f1*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
print(f"Precisão: {precision*100:.2f}%")
```

```python
knn = KNeighborsClassifier(n_neighbors=5)


knn.fit(X_train_scaled, y_train)

predictions = knn.predict(X_test_scaled)
f1 = f1_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')
precision = precision_score(y_test, predictions, average='weighted')

print(f"F1 Score: {f1*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
print(f"Precisão: {precision*100:.2f}%")
```

```python
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)


xgb = XGBClassifier(eval_metric='logloss', random_state=42)
xgb.fit(X_train_scaled, y_train_encoded)


predictions_encoded = xgb.predict(X_test_scaled)


predictions = le.inverse_transform(predictions_encoded)

f1 = f1_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')
precision = precision_score(y_test, predictions, average='weighted')

print(f"F1 Score: {f1*100:.2f}%")
print(f"Recall: {recall*100:.2f}%")
print(f"Precisão: {precision*100:.2f}%")
```

```python
# Escalando os dados com StandardScaler (sem centralizar, pois os dados são esparsos)
scaler = StandardScaler(with_mean=False)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Defina os parâmetros para cada modelo com poucas opções para otimizar a velocidade
param_grid_knn = {
    'n_neighbors': [3, 5],
    'weights': ['uniform'],
    'metric': ['euclidean']
}

param_grid_logreg = {
    'C': [1, 10],
    'penalty': ['l2'],
    'solver': ['lbfgs'],
    'max_iter': [500]
}

param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20],
    'min_samples_split': [2],
    'min_samples_leaf': [1]
}

# Instancie os modelos
knn = KNeighborsClassifier()
logreg = LogisticRegression(max_iter=500)
rf = RandomForestClassifier(random_state=42)

# Configurar o GridSearchCV para cada modelo, priorizando a rapidez
grid_knn = GridSearchCV(knn, param_grid_knn, cv=3, scoring='f1_weighted', n_jobs=-1)
grid_logreg = GridSearchCV(logreg, param_grid_logreg, cv=3, scoring='f1_weighted', n_jobs=-1)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=3, scoring='f1_weighted', n_jobs=-1)

# Treine os modelos usando GridSearchCV
grid_knn.fit(X_train_scaled, y_train)
grid_logreg.fit(X_train_scaled, y_train)
grid_rf.fit(X_train_scaled, y_train)

# Obtenha os melhores parâmetros e melhores scores para cada modelo
best_knn = grid_knn.best_estimator_
best_logreg = grid_logreg.best_estimator_
best_rf = grid_rf.best_estimator_

print("Melhores parâmetros para KNN:", grid_knn.best_params_)
print("Melhores parâmetros para Regressão Logística:", grid_logreg.best_params_)
print("Melhores parâmetros para Random Forest:", grid_rf.best_params_)

# Fazer previsões e calcular métricas para o melhor modelo de cada
for model, name in zip([best_knn, best_logreg, best_rf], ['KNN', 'Logistic Regression', 'Random Forest']):
    predictions = model.predict(X_test_scaled)
    f1 = f1_score(y_test, predictions, average='weighted')
    recall = recall_score(y_test, predictions, average='weighted')
    precision = precision_score(y_test, predictions, average='weighted')

    print(f"\n{name} - F1 Score: {f1*100:.2f}%")
    print(f"{name} - Recall: {recall*100:.2f}%")
    print(f"{name} - Precisão: {precision*100:.2f}%")
```
